```python
import pandas as pd
```

```python
#Step 1: Read the data
df = pd.read_csv(r"E:/Dataset/winequality-red.csv")
df
```

```python
# Step 2: Data Pre-Processing:
# Check the missing values in dataset
df.isnull().sum() #Sum gives you the missing values present in data
```

```python
# Step 3 : Defining X and Y
Y = df[['quality']]
X = df.drop(columns = ['quality'])

# Step 4 : Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8, random_state=1234)
 
len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
# Lets check the number of observations in each category

print("Count of 3 in Y_train", sum(Y_train['quality'] == 3))
print("Count of 4 in Y_train", sum(Y_train['quality'] == 4))
print("Count of 5 in Y_train", sum(Y_train['quality'] == 5))
print("Count of 6 in Y_train", sum(Y_train['quality'] == 6))
print("Count of 7 in Y_train", sum(Y_train['quality'] == 7))
print("Count of 8 in Y_train", sum(Y_train['quality'] == 8))
```

```python
# Import the functionality of SMOTE

from imblearn.over_sampling import SMOTE
```

```python
# We will balance the traing data
# 1. We will create SMOTE object

sm = SMOTE(random_state = 1234)

# 2. We will fit the SMOTE object on X_train, and Y_train --> fit.resample

X_train_new, Y_train_new = sm.fit_resample(X_train,Y_train)
```

```python
print("Count of 3 in Y_train", sum(Y_train_new['quality'] == 3))
print("Count of 4 in Y_train", sum(Y_train_new['quality'] == 4))
print("Count of 5 in Y_train", sum(Y_train_new['quality'] == 5))
print("Count of 6 in Y_train", sum(Y_train_new['quality'] == 6))
print("Count of 7 in Y_train", sum(Y_train_new['quality'] == 7))
print("Count of 8 in Y_train", sum(Y_train_new['quality'] == 8))
```

```python
# Step 4: Creating the model using the training data set

# step a: Create a model object 
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()

# step b: Fit the model object into training data to build a model
model = LR.fit(X_train_new, Y_train_new)
model
```

```python
# Step 5: Predict the values on test data using your model
Y_test['quality'] = model.predict(X_test)
```

```python
Y_test
```

```python
# Step 6: Creating the model using the training data set
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()

model = dtree.fit(X_train_new, Y_train_new)

model
```

```python
# STEP 7: Using this model Predict the X_test Case

Y_test['Pred_Quality'] = model.predict(X_test)

Y_test
```

```python
# STEP 6: To print the classification report

from sklearn.metrics import classification_report

print(classification_report(Y_test['quality'], Y_test['Pred_Quality']))
```
